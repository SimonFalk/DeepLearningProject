```python
import numpy as np
import scipy as sp
import scipy.io as sio
from scipy import signal
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from umap import UMAP
```

```python
files_on  = !ls data/data_tinkhauser_on
files_on  = [f'data/data_tinkhauser_on/{f}'  for f in files_on]
files_off = !ls data/data_tinkhauser_off
files_off = [f'data/data_tinkhauser_off/{f}' for f in files_off]

files = files_on + files_off
```

```python
signals = []
for f in files:
    for i in range(3):
        arr = sio.loadmat(f)['lfp_data'][0][i][0].flatten()
        arr = (arr - arr.mean()) / arr.std()
        signals.append(arr)
        
[
    s.shape
    for s in signals
]
```

```python
def spectrogram(s, fs, window, nperseg, noverlap, nfft, flo, fhi, normalize = True):
    freqs, times, spect = signal.spectrogram(
        s,
        fs,
        window,
        nperseg,
        noverlap,
        nfft,
        scaling = 'spectrum'
    )
    freqs_idxs = (freqs >= flo) & (freqs <= fhi)
    spect = spect[freqs_idxs].T
    
    if normalize:
        spect = np.sqrt(spect)
        spect = (spect - spect.mean(0).reshape((1, -1))) / spect.std(0).reshape((1, -1))
    
    return freqs, freqs_idxs, times, spect

fs = 625
nperseg = 80
noverlap = 75
nfft = 2 ** 12
flo = 8
fhi = 100

windows = [
    'boxcar',
    'triang',
    'blackman',
    'hamming',
    'hann',
    'bartlett',
    'flattop'
]

plt.figure(figsize = (20, 6 * len(windows)))

for idx, window in enumerate(windows):
    freqs, freqs_idxs, times, spect = spectrogram(
        signals[0], 
        fs, 
        window, 
        nperseg, 
        noverlap, 
        nfft, 
        flo, 
        fhi
    )
    freqs = freqs[freqs_idxs]
    
    plt.subplot(len(windows), 1, idx + 1)
    idx_max = (times < 10).sum()
    plt.imshow(
        spect.T[:, :idx_max], 
        aspect = 'auto', 
        interpolation = 'none', 
        cmap = 'gist_heat',
        extent = [
            0, times[idx_max], freqs.max(), freqs.min()
        ]
    )
    plt.title(window)
    plt.xlabel('Time [s]')
    plt.ylabel('Frequency [Hz]')
    
plt.show()
```

```python
X = []
y = []
for idx, s in enumerate(signals):
    freqs, freqs_idxs, times, spect = spectrogram(
        s, 
        fs, 
        window, 
        nperseg, 
        noverlap, 
        nfft, 
        flo, 
        fhi
    )
    y.append([idx] * spect.shape[0])
    X.append(spect)
    
X = np.concatenate(X)
y = np.concatenate(y)
X.shape, y.shape
```

```python
n_components = 10
pca = PCA(n_components).fit(X)
err = ((pca.inverse_transform(pca.transform(X)) - X) ** 2).mean()
X = pca.transform(X)
err, pca.explained_variance_ratio_.cumsum()
```

```python
[(y == idx).sum() for idx in range(18)]
```

```python
t_in  = 3
t_out = 1

X_train = []
y_train = []

for chan in range(18):
    n_rand = int((y == chan).sum() * 3 / (t_in + t_out))
    idxs = np.where(y == chan)[0][:-(t_in + t_out)]
    np.random.shuffle(idxs)
    idxs = idxs[:n_rand]
    
    for idx in idxs:
        X_train.append(X[idx : idx + t_in])
        y_train.append(X[idx + t_in : idx + t_in + t_out])
        
X_train = np.stack(X_train)
y_train = np.stack(y_train)
X_train.shape, y_train.shape
```

```python
model = keras.Sequential([
    layers.Input((t_in, n_components)),
    layers.LSTM(
        100
    ),
    layers.BatchNormalization(),
    layers.Dense(
        50, 
        activation = 'selu', 
    ),
    layers.BatchNormalization(),
    layers.Dense(t_out * n_components),
    layers.Reshape((t_out, n_components))
])

model.compile(
    optimizer ='adam',
    loss = 'mse'
)
model.summary()
```

```python
hist = model.fit(
    X_train, y_train,
    batch_size = 1024,
    epochs = 100,
    validation_split = 0.2
)
```

```python
rows = 4
cols = 8

plt.figure(figsize = (5 * cols, 5 * rows))

axes = [
    (
        plt.subplot(rows, cols, 2 * idx + 1),
        plt.subplot(rows, cols, 2 * idx + 2)
    )
    for idx in range(int(rows * cols / 2))
]

idxs = np.arange(len(y_train))
np.random.shuffle(idxs)
idxs = idxs[:int(rows * cols / 2)]

P = model(X_train[idxs]).numpy()

for (ax1, ax2), idx, p, truth in zip(
    axes,
    idxs,
    P,
    y_train[idxs]
):
#     ax1.imshow(pca.inverse_transform(p).T, aspect = 'auto', interpolation = 'none', cmap = 'gist_heat')
#     ax2.imshow(pca.inverse_transform(truth).T, aspect = 'auto', interpolation = 'none', cmap = 'gist_heat')

    a1 = pca.inverse_transform(p).flatten()
    a2 = pca.inverse_transform(truth).flatten()

    ax1.plot(a1)
    ax1.plot(a2)
    ax2.plot(p.flatten())
    ax2.plot(truth.flatten())

plt.show()
```
